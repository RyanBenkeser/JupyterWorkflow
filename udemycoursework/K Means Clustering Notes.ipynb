```python
# K Means Clustering is an unsupervised learning algorithm
# It attempts to group similar clusters together in your data
# A typical clustering problem clusters similar documents, similar customers, markets and identifying similar physical groups
```

```python
# The overall goal is to divide data into distinct groups such that observations within each group are similar
```

```python
#The K Means algorithm...
#* Chooses a number of clusters "K"
#*Randomly assigns each point to a cluster
#*Until cluster stops changing, repeats the following:
#*For each cluster, compute the cluster centroid by taking the mean vector of points in the cluster
#*Assigns each data point to the cluster for which the centroid is the closest
```

```python
#Choosing a K Value is challenging...  Use the elbow method.
#First, compute the sum of squared error (SSE) for some values of k (for example 2,4,6,8,etc.)
#The SSE is defined as the sum of the squared distance between each member of the cluster and its centroid
#If you plot k against the SSE, you will see that the error decreass as k gets largers; this is because when the number of clusters increass, they should be smaller, so distortion is smaller
#The idea of the elbow method is to choose the k at which the SSE decreases abrubtly.  
```

# K Means Clustering with Python

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples = 200,n_features = 2,centers=4,cluster_std=1.8,random_state=101)
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
data[1]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig , (ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')


```
